```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import BSpline
```

```python
# Define the knot vector
# knots = np.array([0, 0, 0, 0, 1, 2, 3, 4, 4, 4, 4])
knots = np.array([-3, -2, -1, 0, 1, 2, 3, 4, 5, 6, 7])

# Define the degree of the spline
degree = 3

# Create the basis functions
num_basis = len(knots) - degree - 1
basis_funcs = []
x = np.linspace(0, 4, 100)
for i in range(num_basis):
    coeff = np.zeros(num_basis)
    coeff[i] = 1
    spline = BSpline(knots, coeff, degree)
    basis_funcs.append(spline(x))

# Plot the basis functions
plt.figure(figsize=(10, 6))
for i, y in enumerate(basis_funcs):
    plt.plot(x, y, label=f'B_{i,3}(x)')
plt.title('Cubic B-spline Basis Functions')
plt.xlabel('x')
plt.ylabel('B_{i,3}(x)')
plt.legend()
plt.show()
```
